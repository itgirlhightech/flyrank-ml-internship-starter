 # 1. My lane as an ML task
    This lane focuses on predicting an Opportunity Score for each content page based on metrics such as CTR, average position, impressions, content age, and trend.
    The predicted score helps rank pages by their refresh potential, allowing content and SEO teams to prioritize which pages should be updated first.

 # 2. Target or Proxy
    The target is a proxy called Opportunity Score. It represents the potential value of refreshing a content page based on its performance metrics.

 # 3. Success Metrics
    The success metric should measure how well the model prioritizes the best refresh opportunities. A good model should rank the most valuable content higher, helping the team focus on pages with the greatest potential impact.

```python
import pandas as pd

df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

# Show only the most important columns for the lane

df[
    ["content_id",
     "ctr",
     "avg_position", 
     "impressions_90d",
     "days_since_last_update",
     "trend_direction"]
].head()
```

# 4. Unit of analysis, as a real dataframe
     Each content_id represents a unique piece of content. Each row contains the performance metrics for that content,
     and these metrics can be used as features for the model to learn patterns related to update opportunities.

# 5. Why ML beats a fixed rule?
     Machine Learning is more effective than a fixed rule because content performance depends on multiple factors, not just a single condition.
     Metrics such as CTR, average position, impressions, content age, and trend direction interact in different ways. Instead of relying on a simple rule,an ML model can learn these patterns from the data and provide a more accurate Opportunity Score to help prioritize content updates.
